The assessment table contains the test score, homework score and exam score for each student(student_id) in every subject(subject_id) that the student offers in each term (term_id) for each class(class_id).

I want to perform an aggregation the generates one score for each subject(subject_id) that a student( student_id) offers in each term(term_id) for a each class(class_id)

The aggregation will be sum of the test score mutiplied by 0.2(20 percent), the homework score multiplied by 0.1(10 percent) and the exam multiplied by 0.7(70 percent)

information
There are 3 classes
There are 3 terms in a class except class 6(ss3) with just 1 term
Each student offers 10 subjects per term

* Drop the assessment_id column
* pivot the dataframe to create a separate test, homework and exam columns
* Group the data
* Apply the weighted sum

```python
#Drop the assessment_id column
dfa = dfa.drop(columns=['id'])
dfa
```

```python
# pivot the dataframe to create a separate test, homework and exam columns
dfa = dfa.pivot_table(index=['student_id','class_id','term_id','subject_id'],
                                          columns='assessment_type',
                                          values='score').reset_index()
```

```python
dfa.head(50)
```

```python
#checking for null value_counts
dfa.isnull().value_counts()
```

```python
# Calculate the final score using the weighted formula
dfa['final_score'] = (dfa['test'] * 0.2) + (dfa['homework'] * 0.1) + (dfa['exam'] *0.7)
```

```python
dfa
```

```python
#droping the exam, homework and test columns
dfa.drop(columns=['exam', 'homework', 'test'], inplace= True)
dfa
```

```python
#pivoting the subject_id column 
dfa = dfa.pivot_table(index=['student_id','class_id','term_id'],
                                          columns='subject_id',
                                          values='final_score').reset_index()
```
